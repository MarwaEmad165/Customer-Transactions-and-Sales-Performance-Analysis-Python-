** Advanced Data Visualization**

line chart Display monthly revenue trends for each product category.

```python
monthly_category_sales = df.groupby(['Month', 'Product_Category'])['Amount_Spent'].sum().unstack()

# Plotting the line chart
plt.figure(figsize=(12, 6))
monthly_category_sales.plot(kind='line')
plt.title('Monthly Revenue Trends for Each Product Category')
plt.xlabel('Month')
plt.ylabel('Total Revenue ($)')
plt.xticks(rotation=45)
plt.legend(title='Category')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Pie Chart: Show the distribution of sales by payment type.

```python
payment_sales = df.groupby('Payment_Type')['Amount_Spent'].sum()

plt.figure(figsize=(6, 6))
payment_sales.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='Set3')
plt.title('Sales Distribution by Payment Type')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

- Histogram: Illustrate the distribution of transaction amounts.


```python
plt.figure(figsize=(6, 6))
sns.histplot(df['Amount_Spent'], bins=10, kde=True, color='red')
plt.title('Distribution of Transaction Amounts')
plt.xlabel('Transaction_Amount')
plt.ylabel('Frequency')
plt.show()
```

- Area Chart: Compare sales performance across different regions over time.

```python
category_sales_area = df.groupby(['Month', 'Product_Category'])['Amount_Spent'].sum().unstack()

# Plotting the area chart
plt.figure(figsize=(8, 8))
category_sales_area.plot(kind='area',cmap='Set3')
plt.title('Sales Performance Across Categories Over Time')
plt.xlabel('Month')
plt.ylabel('Total Revenue ($)')
plt.xticks(rotation=45)
plt.legend(title='Category')
plt.tight_layout()
plt.show()
```

Bonus Challenges
- Calculate customer retention rate based on repeat purchases.

```python
repeat_customers = df.groupby('Cust_Name').size()
retained_customers = repeat_customers[repeat_customers > 1].count()

# Total number of unique customers
total_customers = df['Cust_Name'].nunique()

# Retention rate
retention_rate = retained_customers / total_customers * 100
print(f"Customer Retention Rate: {retention_rate:.2f}%")
```

```python
repeat_customers = df[df.duplicated(subset=['Customer_ID'], keep=False)]
total_customers = df['Customer_ID'].nunique()
retention_rate = len(repeat_customers['Customer_ID'].unique()) / total_customers * 100
print(f"Customer Retention Rate: {retention_rate:.2f}%")
```


- Compare average spending per region.

```python
avg_spending_per_region = df.groupby('Region')['Amount_Spent'].mean()
print("\nAverage Spending Per Region:")
print(avg_spending_per_region)
```


- Analyze the correlation between the number of units purchased and the total amount spent.

```python
correlation = df['Units_Purchased'].corr(df['Amount_Spent'])
print(f"Correlation between Units Purchased and Amount Spent: {correlation:.2f}")

# scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Units_Purchased', y='Amount_Spent', data=df)
plt.title('Scatter Plot: Units Purchased vs Amount Spent')
plt.xlabel('Units Purchased')
plt.ylabel('Amount Spent')
plt.tight_layout()
plt.show()
```


- Create a heatmap using Seaborn to show sales trends by weekday and hour.

```python
df['purchase_date'] = pd.to_datetime(df['purchase_date'])
df['weekday'] = df['purchase_date'].dt.weekday
df['hour'] = df['purchase_date'].dt.hour
sales_by_day_hour = df.groupby(['weekday', 'hour'])['Amount_Spent'].sum().unstack()

# Plotting the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(sales_by_day_hour, cmap='coolwarm', annot=True, fmt='.1f', linewidths=.5)
plt.title("Sales Trends by Weekday and Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Weekday")
plt.show()
```

```python
np.random.seed(42)  # for reproducibility
df['hour'] = np.random.randint(0, 24, size=len(df))

# Set the correct order for weekdays
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['Weekday'] = pd.Categorical(df['Weekday'], categories=weekday_order, ordered=True)

# Pivot table to structure the data for the heatmap
heatmap_data = df.pivot_table(
    index='Weekday',
    columns='hour',
    values='Amount_Spent',  # change this to 'Amount_Spent' from 'Transaction Amount'
    aggfunc='sum',
    fill_value=0
)

# Create the heatmap
plt.figure(figsize=(28, 18))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='.1f', cbar_kws={'label': 'Total Sales'})
plt.title('Sales Trends by Weekday and Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')
plt.tight_layout()
plt.show()
```
